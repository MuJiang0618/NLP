# 文本关键词提取

```python
import jieba
from sklearn.feature_extraction.text import TfidfVectorizer
import re
import numpy as np
```

# 预处理
引入语料库, 去除特殊符号, 并进行分词处理

```python
filename_list = [2,3,5,7,9]
r1 =  '[a-zA-Z0-9’\n·\s＊!"：#$%&\'()◆●（）＠②*+,-./:;<=>?@，。?★、…【】《》？——“”‘’！[\$$^_`{|}~]+'
corpus_list = []

for i in filename_list:   #取了5篇文档, 可以试试取更多
    with open(r'D:\CS\dataset\NLP\answer\C3-Art\C3-Art000'+ str(i) + '.txt', 'r', encoding='utf-8') as f:
        text = f.read()
        text = re.sub(r1, '', text)
        corpus = ' '.join(jieba.cut(text))
        corpus_list.append(corpus)
```

引入停用词

```python
with open(r'D:\Repositories\Datasets\NLP\chineseStopWords.txt', 'r', encoding='utf-8-sig') as f:
    content = f.read()
    stopwordslist = content.splitlines()
```

## TF-IDF

```python
vector = TfidfVectorizer(stop_words=stopwordslist)
tfidf = vector.fit_transform(corpus_list)
```

```python
wordlist = vector.get_feature_names()   # 获取所有词
tfidf_value = tfidf.toarray()
```

```python
wordlist
```

```python
topic_words = []
topN = 10      # 取topN个词
for i in range(tfidf_value.shape[0]):
    topic_words_index = np.argsort(- tfidf_value[i])[:topN]
    temp_list = []
    for j in topic_words_index:
        temp_list.append(wordlist[j])
    topic_words.append(temp_list)
```
