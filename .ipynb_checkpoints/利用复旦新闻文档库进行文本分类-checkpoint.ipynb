# 利用复旦新闻文档库进行文本分类

```python
import numpy as np
import jieba
import re
```

 在 艺术, 历史, 经济, 环境, 运动5个类别各选取10篇文档作为训练集

获取每个文档的词汇列表,再去除每个词汇列表中的停用词,然后合并所有词汇列表得到这50篇文档的词袋

```python
def textParse(text_content):
    r1 = '[a-zA-Z0-9’\n·\s＊!"：#$%&\'()◆●（）*+,-./:;<=>?@，。?★、…【】《》？——“”‘’！[\$$^_`{|}~]+'
    result = re.sub(r1, '', text_content)     #去掉特殊字符
   
    ChineseStopWords = open(r'D:\Repositories\NLP\dataset\chineseStopWords.txt', 'r', encoding='utf-8').read()
    ChineseStopWords = re.sub('\n', '', ChineseStopWords)
    
    return re.sub(ChineseStopWords, '', result)  
```

可以考虑分词后再去掉停止词,这样去停止词的精度更高,因为分词后非单字符停止词会在一起;但是复杂度更高

```python
def getWordList():
    doc_list = [] ; class_list = []
    word_vector = []
    # 由于文件名数字部分非连续且无规律,只能手动选择10篇
    art_numbering = ['0002','0003','0005','0007','0009','0011','0013','0015','0017','0019']  #类别0
    history_numbering = ['001','002','004','006','008','010', '012', '014', '016','018']    #  1
    economy_numbering = ['0002','0003','0004','0006','0008','0010','0012','0014','0016','0018']   # 2
    environ_numbering = ['0002','0003','0005','0007','0009','0011','0013','0015','0017','0019'] # 3
    sports_numbering = ['0001','0003','0005','0007','0009','0011','0013','0015','0017','0019']   # 4
    
    temp_list = []
    temp_list.append(art_numbering) ; temp_list.append(history_numbering); temp_list.append(economy_numbering)
    temp_list.append(environ_numbering) ; temp_list.append(sports_numbering)
                         
    dirlist = ['C3-Art', 'C7-History', 'C34-Economy', 'C31-Enviornment', 'C39-Sports']
    for i in range(len(dirlist)):
        for numbering in temp_list[i]:
            word_list = textParse(open(r'D:\CS\dataset\NLP\answer\%s\%s%s.txt' %(dirlist[i], dirlist[i], numbering), encoding='utf-8').read())  
            class_list.append(i)
            word_vector.append(word_list)
            
    return word_vector, class_list
```

```python
def getWordFrequency():
    word_vector, class_list = getWordList()
    dirlist = ['C3-Art', 'C7-History', 'C34-Economy', 'C31-Enviornment', 'C39-Sports']
    word_freq_dict = {}
    
    for i in range(len(word_vector)):
        for word in jieba.cut(word_vector[i]):
            word_freq_dict.setdefault(word, {})
            word_freq_dict[word].setdefault(dirlist[class_list[i]], 0)
            word_freq_dict[word][dirlist[class_list[i]]] += 1  #这里可以修改使词频字典里的类别为字符串,更直观
            
    return word_freq_dict
```

```python
word_freq_dict = getWordFrequency()
```
